3. Ещё раз посмотрите на долю или количество вхождений каждого значения целевой переменной в выборку и напишите, какое это распределение. Определить, как распределены значения целевой переменной, в данном случае можно и без графика. 

**Ответ:** _переменная "price_category" имеет относительно равномерное распределение между категориями "high", "medium" и "low", без явного доминирования какой-либо одной категории._

````python
import pandas as pd
import matplotlib.pyplot as plt

data = {'price_category': ['high', 'medium', 'low'],
        'count': [0.349652, 0.327761, 0.322587]}

df = pd.DataFrame(data)

plt.figure(figsize=(6, 6))
plt.pie(df['count'], labels=df['price_category'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Price Categories Distribution')
plt.show()
````
